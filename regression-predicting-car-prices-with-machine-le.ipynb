```python
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
from pandas.core.dtypes.common import is_numeric_dtype

for column in df.columns:
    if is_numeric_dtype(df[column]):
        continue
    else : 
        df[column] = lb.fit_transform(df[column])
        
    
```

```python
df.head()
```

```python
df.corr()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# If your dataset is not too big (e.g., < 1000 rows)
sns.pairplot(df, y_vars=['price'], x_vars=df.columns.drop('price'))
plt.show()

```

```python
df = df.drop(['engine-location'],axis =1)
df.head()
```

```python
x = df.drop(['price'],axis = 1)
y = df[['price']]
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,train_size = 0.70,random_state = 42)
```

```python
xtrain.shape
```

```python
ytrain.shape
```

```python
from sklearn.linear_model import LinearRegression
LR = LinearRegression()
LR.fit(xtrain,ytrain)
```

```python
LR.score(xtest,ytest)
```

```python
#logistic
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression()
LR.fit(xtrain,ytrain)
```

```python
LR.score(xtest,ytest)
```

```python
from sklearn.neighbors import KNeighborsRegressor
KNR = KNeighborsRegressor()
KNR.fit(xtrain,ytrain)
```

```python
KNR.score(xtest,ytest)
```

```python
KNR = KNeighborsRegressor(n_neighbors = 20,weights = 'distance')
KNR.fit(xtrain,ytrain)
KNR.score(xtest,ytest)
```

```python
import matplotlib.pyplot as plt

from sklearn.metrics import mean_squared_error
y_pred_knnc = KNR.predict(xtest)
mse = mean_squared_error(ytest,y_pred_knnc)
print(mse)
print("It has almost 80% score means The model correctly classified 80 out of every 100 samples.")


plt.scatter(range(len(ytest)), ytest, label='Actual', alpha=0.6)
plt.scatter(range(len(ytest)), y_pred_knnc, label='Predicted', alpha=0.6)
plt.title("kNN Classifier Predictions")
plt.xlabel("Sample Index")
plt.ylabel("Price")
plt.legend()

plt.tight_layout()
plt.show()
```
